```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("AirQualityUCI.xlsx")
```

```python
from sklearn.preprocessing import MinMaxScaler

# Drop non-numeric columns like 'Date' and 'Time' if they exist and are not needed for scaling
df_numeric = df.drop(columns=['Date', 'Time'], errors='ignore')

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling to the numeric DataFrame
df_scaled = scaler.fit_transform(df_numeric)

# Convert the scaled array back to a DataFrame with original column names
df_scaled = pd.DataFrame(df_scaled, columns=df_numeric.columns)

print("DataFrame after Min-Max Scaling:")
print(df_scaled.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler_standard = StandardScaler()

# Apply Standard Scaling to the numeric DataFrame
df_standard_scaled = scaler_standard.fit_transform(df_numeric)

# Convert the scaled array back to a DataFrame with original column names
df_standard_scaled = pd.DataFrame(df_standard_scaled, columns=df_numeric.columns)

print("DataFrame after Standard Scaling:")
print(df_standard_scaled.head())
```

### Comparison of Transformed Outputs

Let's compare the statistical properties of the data after Min-Max Scaling and Standard Scaling.

```python
print("Statistical Summary after Min-Max Scaling:")
display(df_scaled.describe())

print("\nStatistical Summary after Standard Scaling:")
display(df_standard_scaled.describe())
```
