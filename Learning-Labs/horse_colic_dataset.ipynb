```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("horse colic dataset.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df['rectal_temp'] = df['rectal_temp'].fillna(df['rectal_temp'].median())
df['pain'] = df['pain'].fillna(df['pain'].mode()[0])
df_filled = df.fillna('Unknown')
```

```python
df['pulse'] = df['pulse'].fillna(df['pulse'].mean())
df['respiratory_rate'] = df['respiratory_rate'].fillna(df['respiratory_rate'].mean())
df['nasogastric_reflux_ph'] = df['nasogastric_reflux_ph'].fillna(df['nasogastric_reflux_ph'].mean())
df['packed_cell_volume'] = df['packed_cell_volume'].fillna(df['packed_cell_volume'].mean())
df['total_protein'] = df['total_protein'].fillna(df['total_protein'].mean())
df['abdomo_protein'] = df['abdomo_protein'].fillna(df['abdomo_protein'].mean())

print("Null values after filling numerical columns with mean:")
df.isnull().sum()
```

```python
for col in ['temp_of_extremities', 'peripheral_pulse', 'mucous_membrane', 'capillary_refill_time', 'peristalsis', 'abdominal_distention', 'nasogastric_tube', 'nasogastric_reflux', 'rectal_exam_feces', 'abdomen', 'abdomo_appearance']:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

print("Null values after filling categorical columns with mode:")
df.isnull().sum()
```

### Demonstrating Bias and Variance

Bias and variance are key concepts in machine learning that describe sources of error in a model.

To understand them in practice, we need to:
1.  **Define a task**: What are we trying to predict?
2.  **Split data**: Separate data into training and testing sets.
3.  **Train models**: Train models with varying complexity.
4.  **Evaluate**: Compare training and testing errors to infer bias and variance.

Let's set up a simple regression task: predicting `packed_cell_volume` based on other numerical features.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Define features (X) and target (y)
# We'll use only numerical features that are fully filled after the cleaning steps
features = ['rectal_temp', 'pulse', 'respiratory_rate', 'nasogastric_reflux_ph', 'total_protein', 'abdomo_protein']
target = 'packed_cell_volume'

X = df[features]
y = df[target]

# Drop rows with any remaining NaN values in X or y to ensure clean data for model training
combined_data = pd.concat([X, y], axis=1).dropna()
X = combined_data[features]
y = combined_data[target]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```
