```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mushroom secondary_data_csv.csv", delimiter=';')
```

```python
df.head()
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
missing_values = df.isnull().sum()
missing_values_percentage = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_values_percentage
})

display(missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False))
```

```python

final_missing_check = df.isnull().sum()
final_missing_percentage = (df.isnull().sum() / len(df)) * 100

final_missing_df = pd.DataFrame({
    'Missing Count': final_missing_check,
    'Missing Percentage': final_missing_percentage
})

display(final_missing_df[final_missing_df['Missing Count'] > 90])

if final_missing_df['Missing Count'].sum() == 0:
    print("\nAll missing values have been successfully removed from the DataFrame.")
else:
    print("\nThere are still missing values in the DataFrame. Please review the output above.")
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
print('Original DataFrame info:')
print(df.info())

categorical_cols = df.select_dtypes(include=['object']).columns
print(f"\nCategorical columns identified: {list(categorical_cols)}")
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
print('\nDataFrame info after one-hot encoding:')
print(df_encoded.info())
```
