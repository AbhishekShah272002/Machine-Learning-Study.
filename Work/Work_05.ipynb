```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data = {'names': ['john', 'michael', 'sarah', 'jessica', 'david'],
        'ages': [25, 30, 22, 28, 35],
        'gender': ['male', 'male', 'female', 'female', 'male'],
        'rank': [1, 2, 3, 4, 5]}

df=DataFrame(data)
df
```

Matplotlib's Bar Chart

```python
plt.bar(df['names'], df['ages'])
plt.xlabel('Names')
plt.ylabel('Ages')
plt.title('Comparing Ages')
plt.show() 
```
